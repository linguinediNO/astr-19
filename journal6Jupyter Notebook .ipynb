# Journal 6 aka table of sin(x) and cos(x)
This notebook will define functions to find 'sin(x)' and 'cos(x)', and then create a table of values for x from 0 to 2pi.

We will:
1. Define a function for sin(x)
2. Define a function for cos(x).
3. Tabulate sin(x) and cos(x) for 1000 x values.
4. Use a loop to print the first 10 rows for x, sin(x), and cos(x) in columns. 

```python
import numpy as np

#define functions
def my_sin(x):
    return np.sin(x)
```

```python
def my_cos(x):
    return np.cos(x)
```

```python
#Generate 1000 x values between 0 and 2*np.pi
x = np.linspace(0, 2*np.pi, 1000)

#get sin(x) and cos(x) for each x
sin = my_sin(x)
cos = my_cos(x)
```

```python
#print the first 10 rows 
print("  x       sin(x)       cos(x)")
print("_____________________________")

for i in range(10):
    print(x[i], sin[i], cos[i])
```

This loop prints the first 10 values of x, sin(x), and cos(x) in columns.
