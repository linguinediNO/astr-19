```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
random_numbers = np.random.exponential(scale=1.0, size=1000)
```

```python
plt.figure(figsize=(8,4))

plt.hist(random_numbers, bins=1000, facecolor='skyblue', edgecolor='pink')

plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of Exponential Random Numbers")

plt.savefig("exponential_histogram.pdf")
plt.show()
```
